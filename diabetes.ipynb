```python
from sklearn.model_selection import train_test_split
y=df['Outcome'].values
x=df.drop(["Outcome","outcome_cat"],axis=1).values
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
###Creating Tensors
X_train=torch.FloatTensor(x_train)
X_test=torch.FloatTensor(x_test)
Y_train=torch.LongTensor(y_train)
Y_test=torch.LongTensor(y_test)
```

```python
X_train[:5]
```

```python
Y_train[:5]
```

```python
Y_train.dtype
```

```python
####Creating model with pytorch
class ANN_Model(nn.Module):
    def __init__(self,input_features=8,hidden1=20,hidden2=20,out_features=2):
        super().__init__()
        self.f_connected1 = nn.Linear(input_features,hidden1)
        self.f_connected2 = nn.Linear(hidden1,hidden2)
        self.out = nn.Linear(hidden2,out_features)
        
    def forward(self,x):
        x=F.relu(self.f_connected1(x))
        x=F.relu(self.f_connected2(x))
        return self.out(x)
        
    
```

```python
####Instantiate ANN_Model
random_seed=42
torch.manual_seed(random_seed)
model=ANN_Model()
```

```python
model.parameters
```

```python
###Backward Propagation and optimization
loss_function=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
epochs=500
final_losses=[]
for i in range(epochs):
    i=i+1;
    y_pred=model.forward(X_train)
    loss=loss_function(y_pred,Y_train)
    final_losses.append(loss)
    if i%100==0:
        print(f"Epoch number is:{i} and the total loss is: {loss}")
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    
```

```python
# %Plot the loss function
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plotting the losses over time (epochs/iterations)
final_losses = [loss.detach().numpy() for loss in final_losses]
plt.plot(range(epochs), final_losses)

# Adding labels and title for clarity
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')

# Display the plot
plt.show()
```

```python
predictions=[]
with torch.no_grad():
    for i,x in enumerate(X_test):
        pred=model(x)
        predictions.append(pred.argmax().item())
        print(pred.argmax().item())
```

```python
predictions[:5]
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
c_matrix=confusion_matrix(Y_test,predictions)
```

```python
print(c_matrix)
```

```python
c_report=classification_report(Y_test,predictions)
print(c_report)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(c_matrix,annot=True)
plt.xlabel('Actual_Value')
plt.ylabel('Predicted_Value')
```
